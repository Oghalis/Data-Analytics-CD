```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/vn55le3/Downloads/Data-Analytics-CD/5-Analisis Estadistico/data/titanic.csv')
```

#### Medidas de Tendencia Central: Promedio, Mediana, Moda


```python
print("Medidas de Tendencia Central para 'Age' y 'Fare':")
print("Promedio de Age:", df['Age'].mean())
print("Mediana de Age:", df['Age'].median())
print("Moda de Age:\n", df['Age'].mode())
print("Promedio de Fare:", df['Fare'].mean())
print("Mediana de Fare:", df['Fare'].median())
print("Moda de Fare:\n", df['Fare'].mode())
```

#### Medidas de Dispersión: Varianza, Desviación Estándar, Rango

```python
print("\nMedidas de Dispersión para 'Age' y 'Fare':")
print("Varianza de Age:", df['Age'].var())
print("Desviación Estándar de Age:", df['Age'].std())
print("Rango de Age:", df['Age'].max() - df['Age'].min())

print("Varianza de Fare:", df['Fare'].var())
print("Desviación Estándar de Fare:", df['Fare'].std())
print("Rango de Fare:", df['Fare'].max() - df['Fare'].min())

```

#### Histograma

```python
plt.subplot(1, 2, 1)
sns.histplot(df['Age'].dropna(), bins=20, kde=True, color='blue', alpha=0.7)
plt.title('Distribución de Edades')
plt.xlabel('Edad')  
plt.ylabel('Frecuencia')  

plt.subplot(1, 2, 2)
sns.histplot(df['Fare'], bins=20, kde=True, color='green', alpha=0.7)
plt.title('Distribución del Precio del Billete')
plt.xlabel('Precio del Billete') 
plt.ylabel('Frecuencia') 

plt.tight_layout() 
plt.show()
```

#### Boxplot

```python
plt.subplot(1, 2, 1)
sns.boxplot(y='Age', data=df, color='skyblue')
plt.title('Boxplot de Edades')
plt.ylabel('Edad')  

plt.subplot(1, 2, 2)
sns.boxplot(y='Fare', data=df, color='lightgreen')
plt.title('Boxplot del Precio del Billete')
plt.ylabel('Precio del Billete')  

plt.tight_layout() 
plt.show()
```

```python
mean_age_survived = df[df['Survived'] == 1]['Age'].mean()
mean_age_not_survived = df[df['Survived'] == 0]['Age'].mean()

print("\nAnálisis Estadístico: Edad Media")
print("Edad media de los supervivientes:", mean_age_survived)
print("Edad media de los no supervivientes:", mean_age_not_survived)

diff = mean_age_survived - mean_age_not_survived
print("Diferencia en la edad media:", diff)

plt.figure(figsize=(10, 6))
sns.histplot(df, x='Age', hue='Survived', element='step', stat='density', common_norm=False)
plt.title('Distribución de Edad por Supervivencia')
plt.show()
```
